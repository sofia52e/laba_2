```python
from collections import deque

def topological_sort(v):
    n = len(v)
    g = [[] for a in range(n)]
    indeg = [0] * n           #колво входящих ребер
    for i in range(n - 1):
        g[i].append(i + 1)
        indeg[i + 1] += 1     #создаем граф
    q = deque([i for i in range(n) if indeg[i] == 0])
    order = []
    while q:
        u = q.popleft()
        order.append(u)
        for nei in g[u]:
            indeg[nei] -= 1
            if indeg[nei] == 0:
                q.append(nei)
    return [v[i] for i in order]

def smooth_sort(v):
    v = v[:]  #копия списка
    n = len(v)
    if n < 2:
        return v
    leonardo = [1, 1]
    while leonardo[-1] < n:
        leonardo.append(leonardo[-1] + leonardo[-2] + 1)

    def sift(i, size):
        while size > 1:
            r = i - 1
            l = i - 1 - leonardo[size - 2]
            if v[l] >= v[r]:
                m = l
                size -= 1
            else:
                m = r
                size -= 2
            if v[m] <= v[i]:
                break
            v[i], v[m] = v[m], v[i]
            i = m

    def trinkle(i, p, size):
        while p > 0:
            while p % 2 == 0:
                p //= 2
                size += 1
            j = i - leonardo[size]
            if v[j] <= v[i]:
                break
            v[i], v[j] = v[j], v[i]
            i = j
            p //= 2
            size -= 1
            sift(i, size)

    q = 1
    size = 1
    p = 1
    while q < n:
        if (p & 3) == 3:
            sift(q - 1, size)
            p >>= 2
            size += 2
        else:
            if leonardo[size - 1] >= n - q:
                trinkle(q - 1, p, size)
            else:
                sift(q - 1, size)
            if size == 1:
                p <<= 1
                size -= 1
            else:
                p = (p << 1) + 1
                size -= 2
        q += 1
    size -= 1
    p = (p << 1) + 1
    while size >= 0:
        trinkle(n - 1, p, size)
        n -= 1
        if (p & 1) == 1:
            p >>= 1
            size -= 1
        else:
            p >>= 2
            size += 1
    return v
    
#проверка
if __name__ == "__main__":
    data = [5, 2, 8, 3, 1]
    print("Исходные данные:", data)
    print("Topological sort:", topological_sort(data))
    print("Smoothsort:", smooth_sort(data))
```
